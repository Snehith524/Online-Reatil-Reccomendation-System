```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity

# Load dataset
file_path = "C:/Users/gampa/OneDrive/Desktop/SNEHITH/Plasmid/Major/OnlineRetail.xlsx"
df = pd.read_excel(file_path)

# Drop missing values (to avoid errors)
df = df.dropna(subset=["CustomerID", "StockCode", "Quantity"])

# Convert CustomerID to integer (if it's not already)
df["CustomerID"] = df["CustomerID"].astype(int)
df["StockCode"] = df["StockCode"].astype(str)

# Create a pivot table (Customers vs. Products)
pivot_table = df.pivot_table(index="CustomerID", columns="StockCode", values="Quantity", aggfunc="sum").fillna(0)

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(pivot_table)

# Compute similarity matrix
customer_similarity = cosine_similarity(scaled_data)
customer_similarity_df = pd.DataFrame(customer_similarity, index=pivot_table.index, columns=pivot_table.index)

# Recommendation function
def recommend_products(customer_id, df, pivot_table, similarity_df, top_n=5):
    """
    Given a CustomerID, recommends the top_n most frequently bought products by similar customers.
    Returns product details (StockCode, Description, Price, Country).
    """
    if customer_id not in similarity_df.index:
        return f"CustomerID {customer_id} not found in data."

    # Find similar customers
    similar_customers = similarity_df[customer_id].sort_values(ascending=False)[1:top_n+1].index

    # Get products purchased by similar customers
    similar_customers_products = df[df["CustomerID"].isin(similar_customers)]["StockCode"].value_counts().index[:top_n]

    # Retrieve product details
    recommended_details = df[df["StockCode"].isin(similar_customers_products)][["StockCode", "Description", "UnitPrice", "Country"]].drop_duplicates()

    return recommended_details

# Example Usage
customer_id = 17850  # Replace with an actual CustomerID from your dataset
recommended_details = recommend_products(customer_id, df, pivot_table, customer_similarity_df)

print(f"Recommended products for CustomerID {customer_id}:")
print(recommended_details)
```
